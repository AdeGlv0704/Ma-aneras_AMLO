```python
import os
import re

bloque_actual = 13

# Carpeta original y carpeta de salida
carpeta_original = fr"D:\Transcripciones_Mañaneras\Bloque_{bloque_actual}"
carpeta_limpia = fr"D:\Transcripciones_Mañaneras_Limpias\Bloque_{bloque_actual}"
os.makedirs(carpeta_limpia, exist_ok=True)

# Función de limpieza básica
def limpiar_texto(texto):
    texto = texto.lower()
    texto = re.sub(r"http\S+|www\S+", "", texto)  # quitar URLs
    texto = re.sub(r"[^a-záéíóúüñ\s]", "", texto)  # quitar símbolos
    texto = re.sub(r"\d+", "", texto)  # quitar números
    texto = re.sub(r"\s+", " ", texto).strip()
    return texto

# Procesar todos los archivos .txt dentro del bloque
for archivo in os.listdir(carpeta_original):
    if archivo.endswith(".txt"):
        try:
            ruta_original = os.path.join(carpeta_original, archivo)
            ruta_limpia = os.path.join(carpeta_limpia, archivo)

            with open(ruta_original, "r", encoding="utf-8") as f:
                contenido = f.read()

            # Separar la transcripción del encabezado (si existe)
            partes = contenido.split("TRANSCRIPCIÓN:")
            texto = partes[1] if len(partes) > 1 else contenido

            texto_limpio = limpiar_texto(texto)

            with open(ruta_limpia, "w", encoding="utf-8") as f:
                f.write(texto_limpio)

            print(f" Limpio: {archivo}")
        except Exception as e:
            print(f" Error con {archivo}: {e}")

print(f" Bloque_{bloque_actual} procesado correctamente.")
```
