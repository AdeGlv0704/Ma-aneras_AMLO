```python
import os
import spacy

# Carga el modelo de spaCy para español
nlp = spacy.load("es_core_news_sm")

bloque_actual = 13
carpeta_entrada = fr"D:\Transcripciones_Mañaneras_Limpias\Bloque_{bloque_actual}"
carpeta_salida = fr"D:\Transcripciones_Mañaneras_Lematizadas\Bloque_{bloque_actual}"
os.makedirs(carpeta_salida, exist_ok=True)

# Procesa cada archivo .txt del bloque
for archivo in os.listdir(carpeta_entrada):
    if archivo.endswith(".txt"):
        try:
            ruta_entrada = os.path.join(carpeta_entrada, archivo)
            ruta_salida = os.path.join(carpeta_salida, archivo)
            
            with open(ruta_entrada, "r", encoding="utf-8") as f:
                texto = f.read()
            
            # Procesa el texto con spaCy
            doc = nlp(texto)
            # Lematiza cada token (ignorando los espacios y signos de puntuación)
            lemas = [token.lemma_ for token in doc if not token.is_space and not token.is_punct]
            
            with open(ruta_salida, "w", encoding="utf-8") as f:
                for lema in lemas:
                    f.write(lema + "\n")
            
            print(f" Lematizado: {archivo}")
        except Exception as e:
            print(f" Error con {archivo}: {e}")

print(f" Lematización del Bloque_{bloque_actual} completada.")
```
